```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Belajar Data Analyst Python /12. Matplotlib subplots/Jumlah Perceraian Menurut Provinsi dan Faktor Penyebab Perceraian (perkara), 2024.csv')
df.head()
```

```python
# Identify and count missing values
print(df.isnull().sum())
print('-------------------------------------------------------------------------')

# Summarize descriptive statistics
print(df.describe().T)
print('-------------------------------------------------------------------------')

# Explore unique values in categorical columns
print( df['Provinsi'].unique())
print('-------------------------------------------------------------------------')

df.info()
```

```python
# Menghapus baris 37-44
filter_df = df.iloc[:38]

# mengganti type data
for col in df.columns:
    if col != 'Provinsi':
        df[col] = pd.to_numeric(df[col], errors='coerce')
print("\nData Types:\n", df.dtypes)

# Identify and count missing values
print("\nMissing Values:\n", filter_df.isnull().sum())
print('-------------------------------------------------------------------------')

# Summarize descriptive statistics
print(filter_df.describe().T)
print('-------------------------------------------------------------------------')

# Explore unique values in categorical columns
print( filter_df['Provinsi'].unique())
print('-------------------------------------------------------------------------')

filter_df.info()
# 4. Summarize descriptive statistics
print("\nDescriptive Statistics:\n", filter_df.describe())
print('-------------------------------------------------------------------------')

# 5. Explore unique values in categorical columns
print("\nUnique Provinces:\n", filter_df['Provinsi'].unique())
print('-------------------------------------------------------------------------')

pd.options.display.width = 1000
print("\nDataFrame Setelah Penggantian (Terpisah):")
print(filter_df)
```

```python
# Mengganti NaN dengan 0
df_filled_nan = filter_df.fillna(0)
df_replaced_dots = df_filled_nan.replace('...', 0)
print("\nDataFrame Setelah Penggantian (Terpisah):")
print(df_replaced_dots)
```

```python
# Provinsi di Pulau Jawa
provinsi_jawa = [
    "DKI Jakarta",
    "Jawa Barat",
    "Banten",
    "Jawa Tengah",
    "DI Yogyakarta",
    "Jawa Timur"
]

# Provinsi di Pulau Kalimantan
provinsi_kalimantan = [
    "Kalimantan Barat",
    "Kalimantan Tengah",
    "Kalimantan Selatan",
    "Kalimantan Timur",
    "Kalimantan Utara"
]

# Provinsi di Pulau Sumatera
provinsi_sumatera = [
    "Aceh",
    "Sumatera Utara",
    "Sumatera Barat",
    "Riau",
    "Kepulauan Riau",
    "Jambi",
    "Bengkulu",
    "Sumatera Selatan",
    "Lampung",
    "Kepulauan Bangka Belitung"
]

# Provinsi di Pulau Sulawesi
provinsi_sulawesi = [
    "Sulawesi Utara",
    "Gorontalo",
    "Sulawesi Tengah",
    "Sulawesi Barat",
    "Sulawesi Selatan",
    "Sulawesi Tenggara"
]

# Provinsa di Pulau Barat
provinsi_barat = [
    "Maluku",
    "Maluku Utara",
    "Papua Barat",
    "Papua Barat Daya",
    "Papua",
    "Papua Selatan",
    "Papua Tengah",
    "Papua Pegunungan"
]
```

```python
plt.figure(figsize=(20,20))

ax1 = plt.subplot2grid((2,4), (0,0))
ax2 = plt.subplot2grid((2,4), (0,1))
ax3 = plt.subplot2grid((2,4), (0,2))
ax4 = plt.subplot2grid((2,4), (1,0))
ax5 = plt.subplot2grid((2,4), (1,1))
ax6 = plt.subplot2grid((2,4), (1,2))

df_replaced_dots['Provinsi'] = df_replaced_dots['Provinsi'].astype(str)
ax1.barh(df_replaced_dots['Provinsi'], df_replaced_dots['Fakor Perceraian - Jumlah']/1000, color='blue', label='Jumlah Perceraian di Indonesia')
ax1.barh(df_replaced_dots['Provinsi'], df_replaced_dots['Fakor Perceraian - Mabuk']/1000, color='black', label='Faktor Perceraian')
ax1.set_xlabel('Jumlah Perceraian')
ax1.set_ylabel('Provinsi')
ax1.set_title('Perbandingan Jumlah Perceraian di Indonesia ')
ax1.legend()


ax2.barh(provinsi_jawa, df_replaced_dots.loc[df_replaced_dots['Provinsi'].isin(provinsi_jawa), 'Fakor Perceraian - Jumlah']/1000, color='lightblue')
ax3.barh(provinsi_kalimantan, df_replaced_dots.loc[df_replaced_dots['Provinsi'].isin(provinsi_kalimantan), 'Fakor Perceraian - Jumlah']/1000, color='lightgreen')
ax4.barh(provinsi_sumatera, df_replaced_dots.loc[df_replaced_dots['Provinsi'].isin(provinsi_sumatera), 'Fakor Perceraian - Jumlah']/1000, color='red')
ax5.barh(provinsi_sulawesi, df_replaced_dots.loc[df_replaced_dots['Provinsi'].isin(provinsi_sulawesi), 'Fakor Perceraian - Jumlah']/1000, color='orange')
ax6.barh(provinsi_barat, df_replaced_dots.loc[df_replaced_dots['Provinsi'].isin(provinsi_barat), 'Fakor Perceraian - Jumlah']/1000, color='yellow')

ax2.set_xlabel('Jumlah Perceraian')
ax3.set_xlabel('Jumlah Perceraian')
ax4.set_xlabel('Jumlah Perceraian')
ax5.set_xlabel('Jumlah Perceraian')
ax6.set_xlabel('Jumlah Perceraian')

ax2.set_title('Jumlah Perceraian \n Pulau Jawa')
ax3.set_title('Jumlah Perceraian \n Pulau Kalimantan')
ax4.set_title('Jumlah Perceraian \n Pulau Sumatera')
ax5.set_title('Jumlah Perceraian \n Pulau Sulawesi')
ax6.set_title('Jumlah Perceraian \n Pulau Barat')

plt.tight_layout()
plt.show()
```
